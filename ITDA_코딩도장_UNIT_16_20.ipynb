#UNIT 18 break, continue로 반복 제어하기

break와 continue?

      break:제어 흐름 중단

      continue:제어 흐름 유지, 코드 실행만 건너뜀.

```python
#while에서 break로 끝내기

i = 0
while True: #무한루프
  print(i)
  i += 1    #i 1씩 증가
  if i == 10:
    break
```

```python
#for 에서 break로 끝내기

for i in range(9999):
    print(i)
    if i ==10:
      break
```

```python
#continue로 코드실행건너뛰기
```

```python
#for에서 = 

for i in range(10):
  if i % 2 ==0:   #홀수만 출력됨
    continue      #= 짝수일때 실행 x
  print(i)
```

```python
#while에서 = 

i = 0
while i <10:
  i += 1
  if i %2 == 0: 
    continue #짝수일떄 건너뜀
  print(i)
```

#UNIT 19. 중첩 루프

19.1 중첩 루프 사용하기

```python
for i in range(5):   #바깥쪽 루프
  for j in range(5): #안쪽 루프
    print('j:',j,sep='',end=' ')
  print('i: ',i,'\\n',sep='')
```

19.2 사각형 별 출력

```python
for i in range(5):
  for j in range(5):
    print('#',end='')
  print() #/n 이 들어있다고 봐,줄바꿈용
```

19.2.1 사각형 모양 바꾸기

```python
for i in range(3):
  for j in range(7):
    print('*',end='')
  print()
```

19.3 계단식 별 출력

```python
for i in range(5):    #0-4 5번 반복,세로방향
  for j in range(5):  #= 가로방향
    if j <= i:
      print('*',end='')
  print()
```

19.3.1 대각선 별 출력

```python
for i in range(5):
  for j in range(5):
    if j ==i:
      print('*',end='')
  print()                
  # 따로 없으면 왼쪽으로 몰림, 공백 필요해.
```

```python
for i in range(5):    #세로
  for j in range(5):  #가로
    if j == i:        #i가 가로와 같을때
      print('*',end='')
    else:
      print(' ',end='')
  print()
```

```python
#역삼각

for i in range(5):
  for j in range(5):
    if j<i:
      print(' ',end='')
    else:
      print('*',end='')
  print()
```

19.6 심사문제

```python
h = int(input())

for i in range(h):
  for j in reversed(range(h)):
    if j > i:
      print(' ',end='')
    else:
      print('*',end='')
    if j < i:
      print('*',end='')

  print()
```
